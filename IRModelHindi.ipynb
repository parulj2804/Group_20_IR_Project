```python
X = data['Post']
y = data['Labels Set']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(stop_words='english', ngram_range=(1,5))
X_train_vector = count.fit_transform(X_train)
X_test_vector = count.transform(X_test)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf = TfidfTransformer()

X_train_tfidf = tfidf.fit_transform(X_train_vector)

X_test_tfidf = tfidf.transform(X_test_vector)
```

```python
from keras.models import Model
from keras.layers import LSTM, Activation, Dense, Dropout, Input, Embedding,SpatialDropout1D
from keras.preprocessing.text import Tokenizer
from keras.preprocessing import sequence
from keras.callbacks import EarlyStopping
from keras.models import Sequential
```

```python
max_words = 40000
max_len = 300
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X_train)
sequences = tokenizer.texts_to_sequences(X_train)
sequences_matrix = sequence.pad_sequences(sequences,maxlen=max_len)
```

```python
model = Sequential()
model.add(Embedding(max_words, 100, input_length=max_len))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
model.summary()
model.compile(loss='binary_crossentropy',metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping,ModelCheckpoint

stop = EarlyStopping(
    monitor='val_accuracy', 
    mode='max',
    patience=5
)

checkpoint= ModelCheckpoint(
    filepath='./',
    save_weights_only=True,
    monitor='val_accuracy',
    mode='max',
    save_best_only=True)
```

```python
history = model.fit(sequences_matrix,y_train,batch_size=1024,epochs = 4,
          validation_split=0.2,callbacks=[stop,checkpoint])
```

```python
test = tokenizer.texts_to_sequences(X_test)
testMat = sequence.pad_sequences(test,maxlen=max_len)
```

```python
accr = model.evaluate(testMat,y_test)
```

```python
lstm = model.predict(testMat)
```

```python
import keras
load_model=keras.models.load_model("./HindiModel.h5")
with open('tokenizerHindi.pickle', 'rb') as handle:
    load_tokenizer = pickle.load(handle)
```

```python
import re
import nltk
nltk.download('stopwords')
from inltk.inltk import tokenize
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))

hindi_text = input()

def preprocessing(text):
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    text = tokenize(hindi_text, "hi")
    text=" ".join(text)
    return text
HText = [preprocessing(hindi_text)]
seq = load_tokenizer.texts_to_sequences(HText)
padded = sequence.pad_sequences(seq, maxlen=300)
pred = load_model.predict(padded)
print("pred", pred)
if pred <= 0.4:
    print("No Hate")
else:
    print("Hate and Abusive")
```
