```python
import pandas as pd
import tweepy

def searchTweets(words, date_since, numtweet):
        db = pd.DataFrame(columns=['username',
                                   'tweet'])
        words = str(words)
        query_terms = words.split('OR')
        tweets_list = []
        for query in query_terms:
          tweets = tweepy.Cursor(api.search,
                                query, lang="hi",
                                since_id=date_since,
                                tweet_mode='extended').items(numtweet)                    
          tweets_list += [tweet for tweet in tweets]
          i = 1
          for tweet in tweets_list:
                  username = tweet.user.screen_name
                  try:
                          text = tweet.retweeted_status.full_text
                  except AttributeError:
                          text = tweet.full_text
                  ith_tweet = [username,text]
                  db.loc[len(db)] = ith_tweet
                  i = i+1
        # filename = str(''.join(words)) + '.csv'
        filename = 'final1.csv'
        db.to_csv('/content/drive/MyDrive/IR_PROJECT/files/' + filename)
 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X ,y, random_state = 42)
```

```python
from keras.models import Model
from keras.layers import LSTM, Activation, Dense, Dropout, Input, Embedding,SpatialDropout1D
from keras.preprocessing.text import Tokenizer
from keras.preprocessing import sequence
from keras.callbacks import EarlyStopping
from keras.models import Sequential
```

```python
max_words = 40000
max_len = 300
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X_train)
sequences = tokenizer.texts_to_sequences(X_train)
sequences_matrix = sequence.pad_sequences(sequences,maxlen=max_len)
```

```python
sequences_matrix
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(sequences_matrix,y, test_size = 0.20, random_state = 42)
```

```python
tr_model = Sequential()
tr_model.add(Embedding(max_words, 100, input_length=max_len))
tr_model.add(SpatialDropout1D(0.20))
tr_model.add(LSTM(100))
tr_model.add(Dense(1, activation='sigmoid'))
tr_model.summary()
tr_model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping,ModelCheckpoint

stop_pt = EarlyStopping(
    monitor='val_accuracy', 
    mode='max',
    patience=5
)
```

```python
history = tr_model.fit(sequences_matrix,y_train,batch_size=1024,epochs = 4,validation_split=0.20,callbacks=[stop_pt])
```

```python
test = tokenizer.texts_to_sequences(X_test)
testMat = sequence.pad_sequences(test,maxlen=max_len)
```

```python
accr = tr_model.evaluate(testMat,y_test)
```

```python
history.history
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
x = range(1, len(acc) + 1)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, acc, 'b', label='Training acc')
plt.plot(x, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(x, loss, 'b', label='Training loss')
plt.plot(x, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
```

```python
import keras
load_model=keras.models.load_model("./HindiModel.h5")
with open('tokenizerHindi.pickle', 'rb') as handle:
    load_tokenizer = pickle.load(handle)
```

```python
import re
import nltk
nltk.download('stopwords')
from inltk.inltk import tokenize
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))

hindi_text = input()

def preprocessing(text):
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    text = tokenize(hindi_text, "hi")
    text=" ".join(text)
    return text
HText = [preprocessing(hindi_text)]
seq = load_tokenizer.texts_to_sequences(HText)
padded = sequence.pad_sequences(seq, maxlen=300)
pred = load_model.predict(padded)
print("pred", pred)
if pred <= 0.56:
    print("No Hate")
else:
    print("Hate and Abusive")
```
